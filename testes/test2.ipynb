```python
import paho.mqtt.client as mqtt
import time
import json

# Configurações do MQTT
MQTT_BROKER = 'www.maqiatto.com'
MQTT_PORT = 1883
MQTT_USER = 'lucasl050503@gmail.com'
MQTT_PASSWORD = '123456'

# Função para quando o cliente se conectar
def on_connect(client, userdata, flags, rc):
    print("Conectado com sucesso. Código de resultado:", rc)

# Configuração do cliente MQTT para publicação
client = mqtt.Client()
client.username_pw_set(MQTT_USER, MQTT_PASSWORD)
client.on_connect = on_connect

# Conectar ao broker MQTT
client.connect(MQTT_BROKER, MQTT_PORT, 60)

# Loop para manter o cliente em execução
client.loop_start()

# Função para criar uma mensagem JSON
def create_message(sensor_id, data_type, data_interval):
    message = {
        "machine_id": "machine_123",  # ID da máquina de exemplo
        "sensors": [
            {
                "sensor_id": sensor_id,
                "data_type": data_type,
                "data_interval": data_interval
            }
        ]
    }
    return json.dumps(message)

# Publicar mensagens nos tópicos
try:
    while True:
        # Mensagem para o tópico de umidade
        moisture_message = create_message("sensor_1", "moisture", 10)
        client.publish("lucasl050503@gmail.com/UPI1/moisture", moisture_message)
        print(f"Mensagem publicada no tópico 'moisture': {moisture_message}")
        
        # Mensagem para o tópico de temperatura
        temperature_message = create_message("sensor_2", "temperature", 10)
        client.publish("lucasl050503@gmail.com/UPI1/temperature", temperature_message)
        print(f"Mensagem publicada no tópico 'temperature': {temperature_message}")

        # Atraso entre as publicações
        time.sleep(10)

except KeyboardInterrupt:
    print("Interrompido pelo usuário")

finally:
    client.loop_stop()
    client.disconnect()
```

```python
import paho.mqtt.client as mqtt
import time
import psutil
import json

# Configurações do MQTT
MQTT_BROKER = 'www.maqiatto.com'
MQTT_PORT = 1883
MQTT_USER = 'warleyxavier.fernandes@gmail.com'
MQTT_PASSWORD = 'wZn&y6Pt'

# Função para quando o cliente se conectar
def on_connect(client, userdata, flags, rc):
    print("Conectado com sucesso. Código de resultado:", rc)

# Configuração do cliente MQTT para publicação
client = mqtt.Client()
client.username_pw_set(MQTT_USER, MQTT_PASSWORD)
client.on_connect = on_connect


client.connect(MQTT_BROKER, MQTT_PORT, 60)

client.loop_start()

try:
    while True:

        cpu_frequency = psutil.cpu_freq().current
        cpu_usage = psutil.cpu_percent(interval=0.5)
        memory_usage = psutil.virtual_memory().percent
        n_input = psutil.net_io_counters().bytes_recv
        n_output = psutil.net_io_counters().bytes_sent  

        client.publish("warleyxavier.fernandes@gmail.com/device_1/cpu_frequency", cpu_frequency)
        client.publish("warleyxavier.fernandes@gmail.com/device_1/cpu_usage", cpu_usage)
        client.publish("warleyxavier.fernandes@gmail.com/device_1/memory_usage", memory_usage)
        client.publish("warleyxavier.fernandes@gmail.com/device_1/network_io", json.dumps({"n_input": n_input, "n_output": n_output}))

        print("Foi")
        time.sleep(5)

except KeyboardInterrupt:
    print("Interrompido pelo usuário")

finally:
    client.loop_stop()
    client.disconnect()
```
